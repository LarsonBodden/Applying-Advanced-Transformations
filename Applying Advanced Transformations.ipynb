```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
df = pd.read_csv('Data/superhero_info.csv')
df.head()
```

```python
df2 = pd.read_csv('Data/superhero_powers.csv')
df2.head()
```

```python
df.info()
```

```python
## examining a single value from the measurements col
mea = df.loc[0,"Measurements"]
print(type(mea))
mea
```

```python
mea = mea.replace("'",'"')
mea
```

```python
## now we can use json.loads
fixed_mea = json.loads(mea)
print(type(fixed_mea))
fixed_mea
```

```python
## use .str.replace to replace all single quotes
df['Measurements'] = df['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
df['Measurements'] = df['Measurements'].apply(json.loads)
df['Measurements'].head()
```

```python
height_weight = df['Measurements'].apply(pd.Series)
height_weight
```

```python

df = pd.concat((df, height_weight), axis = 1)
df.head(2)
```

```python
df = df.drop(columns=['Measurements'])
```

```python
df['Height'] = df['Height'].str.replace('cm', "")
```

```python
df['Weight'] = df['Weight'].str.replace('kg', "")
```

```python
df['Height'] = pd.to_numeric(df['Height'])
```

```python
df['Weight'] = pd.to_numeric(df['Weight'])
```

```python
df.head()
```

```python
df['Hero|Publisher'].head(2)
```

```python
## adding expand=True
df['Hero|Publisher'].str.split('|',expand=True)
```

```python
df[['Hero', 'Publisher']] = df['Hero|Publisher'].str.split('|',expand=True)
```

```python
df = df.drop(columns=['Hero|Publisher'])
```

```python
df.head()
```

```python
df2.head()
```

```python
## adding expand=True
df2['Powers'].str.split(',',expand=False)
```

```python
df2['Powersplit'] = df2['Powers'].str.split(',',expand=False)
```

```python
exploded = df2.explode('Powersplit')
exploded
```

```python
cols_to_make = exploded['Powersplit'].dropna().unique()
cols_to_make
```

```python
for col in cols_to_make:
    df2[col] = df2['Powers'].str.contains(col)
df2.head()
```

```python
final_df = pd.merge(df, df2, left_on="Hero", right_on="hero_names")
final_df
```

```python
# Compare the average weight of super powers who have Super Speed to those who do not.
final_df.groupby('Super Speed')['Weight'].mean()
```

```python
# What is the average height of heroes for each publisher?
final_df.groupby('Publisher')['Height'].mean()
```
